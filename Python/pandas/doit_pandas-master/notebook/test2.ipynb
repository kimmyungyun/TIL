tips 데이터에서 size가 3 ~ 5인 데이터에 대해, time별 tiprate의 평균은?
- 단, tiprate는 tip/total_bill이다.
- 단, 답과 소스를 제출한다.

```python
import seaborn as sns
import pandas as pd

tips = sns.load_dataset("tips")
```

```python
filter_tips = tips[((tips['size']>=3) & (tips['size']<=5)) ] # size 값이 3~5인 데이터만을 필터링
```

```python
filter_tips
```

```python
filter_tips.describe() #size 값의 min이 3이고, max가 5인 것을 봐서 필터링이 잘된 걸 볼 수 있다
```

각 행에 대한 tiprate를 계산했다

```python
filter_tips['tiprate'] = filter_tips['tip'] / filter_tips['total_bill']
```

```python
filter_tips
```

계산된 tiprate의 평균을 구하기 위해 groupby를 진행하고, mean 함수를 사용했다

```python
filter_tips.groupby('time').mean()
```

Lunch의 tiprate 평균 = 0.146827 
Dinner의 tiprate 평균 = 0.148905
