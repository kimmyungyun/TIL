```python
import tensorflow as tf
import numpy as np
```

```python
tf.enable_eager_execution()
tf.__version__
```

```python
X = np.array([1, 2, 3], dtype="float32")
X
```

```python
y = np.array([2.5, 5, 4], dtype="float32")
y
```

```python
W = tf.Variable([2], dtype="float32")
W
```

```python
b = tf.Variable([1], dtype="float32")
b
```

```python
hypothesis = X*W+b
hypothesis
```

```python
hypothesis-y
```

```python
tf.square(hypothesis - y)
```

```python
tf.reduce_mean(tf.square(hypothesis - y))
```

```python
learning_late = 0.1
for i in range(1000):
    with tf.GradientTape() as tape:
        hypothesis = X * W + b
        print(f"hypothesis = {hypothesis}")
        cost = tf.reduce_mean(tf.square(hypothesis - y))
        print(f"cost = {cost}")
    w_grad, b_grad = tape.gradient(cost, [W, b])
    print(f"W_grad : {w_grad}, b_grad : {b_grad}")
    
    W.assign_sub(w_grad * learning_late)
    
    b.assign_sub(b_grad * learning_late)
    
    print(f" W : {W}, b : {b}")
    print("="*50)
```
