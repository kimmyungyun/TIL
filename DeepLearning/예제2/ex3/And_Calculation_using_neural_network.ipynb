## 인공 신경망을 이용한 AND 연산 예시

```python
import numpy as np
import tensorflow as tf
```

## 텐서플로우가 기존 그래프 기반 방식에서 벗어나 그래프 생성 없이 연산을 즉시 실행하는 명령형 프로그래밍 환경을 뜻

## 텐서플로우의 기존 방식은 그래프(텐서의 연산 흐름)을 구성한 후 sess.run() 함수를 통해서 연산이 수행되는 방식인데 이 방식을 제거하고, python과 같은 즉시 실행 방식을 사용하는 것

```python
tf.enable_eager_execution()
```

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype="float32")
X
```

```python
y = np.array([[0], [0], [0], [1]], dtype="float32")
y
```

```python
w = np.array([[1], [2]], dtype="float32")
w
```

```python
b = np.array([3], dtype="float32")
b
```

```python
X.dot(w)
```

```python
def sigmoid(h):
    return 1/(1+np.exp(-h))
```

```python
hypothesis  = sigmoid(X.dot(w) + b)
hypothesis
```

```python
hypothesis-y
```

```python
learning_rate = 0.1
for i in range(100):
    for j in range(2):
        hypothesis = sigmoid(np.dot(X, w) + b)
        print(f'hypothesis = {hypothesis}')
        cost = -1/4 * (y * np.log(hypothesis) + (1 - y)*np.log(hypothesis))
        print(f'cost = {cost}')
        
        X_col_2d = X.T[j].reshape(1, -1)
        w_grad = 1/4 * np.sum(np.dot(X_col_2d, hypothesis-y))
        print(f"w{j}_grad : {w_grad}")
        
        w[j] -= learning_rate*w_grad
        print(f"w{j} : {w[j]}")
        
        
        b_grad = 1/4 * np.sum(hypothesis-y)
        print(f"b{j}_grad : {b_grad}")
        b -= learning_rate * b_grad
        print(f"b{j} : {b}")
```

```python
predict = np.dot(X, w) + b
predict
```

## 위의 것의 값이 잘 안나오는데 그 이유가 sigmoid를 안써서
