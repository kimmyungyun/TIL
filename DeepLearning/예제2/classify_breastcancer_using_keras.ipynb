```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import tensorflow as tf
import numpy as np
```

```python
data = datasets.load_breast_cancer()
data
```

```python
X = data["data"]
X
```

```python
y = data["target"]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

## 아래의 작업을 하는 이유

값을 표준화 시켜서 값의 차이를 줄일 것은 줄이고 키울 것은 키우기 위해

값의 차이가 크지만, 중요하지 않은 것이 있을 수 있고

값의 차이가 작지만 중요한 것이 있을 수 있으므로 그것을 인식시키기 위해

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_train
```

## 딥러닝 모델 초기화

```python
model = Sequential()
```

## Dense 레이어 추가

```python
model.add(Dense(1, input_dim=30, activation="sigmoid"))
```

## 모델 구성 확인

```python
model.summary()
```

## 모델 학습 방법 및 loss 계산 설정

```python
model.compile(loss="binary_crossentropy", optimizer=Adam(lr=0.001), metrics=['acc'])
```

## 모델 학습 시작

```python
model.fit(X_train, y_train, epochs=1000)
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
pred = model.predict(X_test)
pred
```

```python
pred01 = np.where(pred>0.5, 1, 0)
pred01
```

```python
pred02 = pred01.flatten()
pred02
```

```python
pred03 = (pred02 == y_test)
pred03
```

```python
acc = np.sum(pred03) / len(pred03)
acc
```
