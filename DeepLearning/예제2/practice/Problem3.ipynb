```python
from sklearn.datasets import load_iris
import numpy as np
```

## 데이터 로드

```python
data = load_iris()
data
```

```python
X = data.data
X
```

## y 데이터 원핫 인코딩 수행
## 이유는 모르겠으나 OneHotEncoder를 통해서 transform 할 시에 리턴 값이 scipy의 sparse matrix로 나와서 .toarray() 함수를 사용해줌

```python
y = data.target
y = y.reshape(-1, 1)
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
enc.fit(y)
y = enc.transform(y)
y = np.array(y.toarray(), dtype="float32")
y
```

## 훈련용 데이터와 테스트용 데이터 분햘

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

## 훈려용 데이터 스케일러 적용

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_train
```

## 모델 구성

```python
model = Sequential()
model.add(Dense(16, input_dim=4, activation="sigmoid"))
model.add(Dense(3, activation="softmax"))
model.summary()
```

## 학습 방법, cost 함수 및 학습률 0.01로 세팅

```python
optimizers = Adam(0.001)
model.compile(loss="categorical_crossentropy", optimizer=optimizers, metrics=["acc"])
```

## 학습 수행

```python
model.fit(X_train, y_train, epochs=1000)
```

## 테스트 데이터 스케일러 적용

```python
X_test = scaler.transform(X_test)
p = model.predict(X_test)
p1 = np.argmax(p, 1)
p1
```

```python
target = np.argmax(y_test, 1)
target = target.reshape(-1)
target
```

```python
pred2 = (p1 == target)
pred2
```

## 정확도 계산

```python
acc = np.sum(pred2) / len(pred2)
acc
```
