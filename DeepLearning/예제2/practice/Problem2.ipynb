```python
import tensorflow as tf
import numpy as np
from sklearn.datasets import load_iris
```

```python
tf.enable_eager_execution()
```

## 데이터 로드

```python
data = load_iris()
data
```

```python
X = data.data
X = np.array(X, dtype="float32")
X
```

```python
y = data.target
y = np.array(y, dtype="float32")
y = y.reshape([-1, 1])
y
```

## 트레인셋과 테스트셋으로 분할

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## 이유는 모르겠으나 OneHotEncoder를 통해서 transform 할 시에 리턴 값이 scipy의 sparse matrix로 나와서 .toarray() 함수를 사용해줌

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoder.fit(y_train)
onehot_y_train = encoder.transform(y_train)
onehot_y_train = np.array(onehot_y_train.toarray(), dtype="float32")
onehot_y_train
```

```python
print(f"X_train size : {X_train.shape}, y_train size : {onehot_y_train.shape}")
print(f"X_test size : {X_test.shape}, y_test size : {y_test.shape}")
```

## 훈련데이터 StandardScaler 적용

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
```

## weight 값 초기화 객체 생성

```python
from tensorflow.contrib.layers import xavier_initializer
Initializer = xavier_initializer()
```

## 모델 weight 값 변수 생성 및 초기화

```python
w0 = tf.Variable(Initializer([4, 20]))
w0
```

```python
b0 = tf.Variable(Initializer([ 20]))
b0
```

```python
w1 = tf.Variable(Initializer([20, 3]))
w1
```

```python
b1 = tf.Variable(Initializer([3]))
b1
```

## 훈련 Adam 객체 생성 학습률 0.005로 셋팅

```python
optimizer = tf.train.AdamOptimizer(0.005)
```

## 훈련 수행

```python
for epochs in range(1000):
    with tf.GradientTape() as tape:
        hypothesis0 = tf.sigmoid(tf.matmul(X_train, w0) + b0)
        hypothesis1 = tf.nn.softmax(tf.matmul(hypothesis0, w1) + b1)

        cost = -tf.reduce_sum(onehot_y_train * tf.log(hypothesis1)) /  hypothesis1.get_shape().as_list()[0]
        grads = tape.gradient(cost, [w0, w1, b0, b1])

        optimizer.apply_gradients(grads_and_vars=zip(grads, [w0, w1, b0, b1]))
        if epochs % 100 == 0:
            print("=" * 50)
            print(f"epoch : {epochs}, cost : {cost.numpy()}")
```

## 테스트용 X데이터 scaler 적용

```python
X_test = scaler.transform(X_test)
X_test
```

## 훈련된 모델을 활용한 예측

```python
hypothesis = tf.sigmoid(tf.matmul(X_test, w0) + b0)
pred = tf.nn.softmax(tf.matmul(hypothesis, w1) + b1)
pred
```

```python
pred02 = tf.argmax(pred, 1)
pred02
```

## 정확도 측정

```python
acc = tf.reduce_sum(tf.cast(tf.equal(pred02, y_test.reshape(-1)), dtype="float32")) / len(y_test)
acc
```
