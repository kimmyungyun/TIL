## 공분산

$$
Cov_{xy} = \frac{\sum{((x-\bar{x}) (y-\bar{y}))}}{n-1}
$$

## 상관관계
- 공분산을 X와 Y의 표준편차(X,Y에 저장된 값들의 차이)로 나눠서 비율을 조회
- 숫자값이 저장된 컬럼 사이의 상관 관계를 계산

## 상관계수
- 상관계수 = 공분산 / 표준편차
- 표준편차가 작으면 밀도가 촘촘해진다
- 밀도가 촘촘하면 상관계수가 크다

## 상관관계 분석
- 두 변수간에 어떤 선형적 관계를 갖고 있는지를 분석하는 방법
- 연속형 수치로 이루어진 변수와 변수 사이의 관계를 말함
