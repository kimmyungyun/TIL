```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
data = datasets.load_iris()
data
```

```python
X = data.data
X = np.array(X, dtype="float32")
X
```

```python
y = data["target"]
y = np.array(y, dtype="float32")
y = y.reshape(-1, 1)
y
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
enc.fit(y)
```

```python
y = enc.transform(y).toarray()
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_train
```

```python
model = Sequential()
```

```python
model.add(Dense(16, input_dim=4, activation="sigmoid"))
```

```python
model.add(Dense(3, input_dim=16, activation="softmax"))
```

```python
model.summary()
```

```python
optimizers = Adam(0.001)
model.compile(loss="categorical_crossentropy", optimizer=optimizers, metrics=["acc"])
```

```python
model.fit(X_train, y_train, epochs=1000)
```

```python
X_test = scaler.transform(X_test)
p = model.predict(X_test)
```

```python
p1 = np.argmax(p, 1)
p1
```

```python
target = np.argmax(y_test, 1)
target
```

```python
pred2 = (p1 == target)
pred2
```

```python
np.sum(pred2) / len(pred2)
```
