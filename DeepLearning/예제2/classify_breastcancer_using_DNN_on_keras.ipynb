```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

## 데이터 로드

```python
data = datasets.load_breast_cancer()
data
```

## 데이터 X와 y로 나누기

```python
X = data['data']
X = np.array(X, dtype="float32")
X
```

```python
y = data['target']
y = np.array(y, dtype="float32")
y = y.reshape(-1, 1)
y
```

## 훈련용, 테스트용으로 나누기

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

## 훈련용 데이터 표준화

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_train
```

## 모델 정의

```python
model = Sequential()
```

```python
model.add(Dense(30, input_dim = 30, activation="sigmoid"))
```

```python
model.add(Dense(10, input_dim = 30, activation="sigmoid"))
```

```python
model.add(Dense(1, input_dim = 10, activation="sigmoid"))
```

## 모델 확인

```python
model.summary()
```

## 학습률 정의

```python
lr = 0.01
```

## 학습 파라미터 초기화

```python
model.compile(loss="binary_crossentropy", optimizer = Adam(lr), metrics=['acc'])
```

## 학습 시작

```python
model.fit(X_train, y_train, epochs=1000)
```

## 테스트용 데이터 표준화

```python
X_test = scaler.transform(X_test)
X_test
```

## 테스트용 데이터를 활용한 예측

```python
pred = model.predict(X_test)
pred
```

```python
pred01 = np.where(pred>0.5, 1, 0)
pred01
```

## 정확도 측정

```python
pred02 = (pred01==y_test)
acc = np.sum(pred02)/len(pred02)
acc
```
