```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import numpy as np
```

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
], dtype="float32")
X
```

```python
y = np.array([0, 0, 0, 1], dtype="int32")
y
```

## 케라스 모델 초기화

```python
model = Sequential()
```

## 케라스 레이어 추가(인풋 2, 아웃풋 1인 레이어)

```python
model.add(Dense(1, input_dim=2, activation="sigmoid"))
```

```python
model.layers[0].get_weights()
```

```python
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer=Adam(lr=0.1), metrics=['acc'])
```

```python
model.fit(X, y, epochs=500)
```

```python
pred = model.predict(X)
pred
```

```python
predict01 = np.where(pred>0.5, 1, 0)
predict01
```

```python
predict02 = predict01.flatten()
predict02
```

```python
predict03 = (predict02 == y)
predict03
```

```python
acc = np.sum(predict03)/4
acc
```
