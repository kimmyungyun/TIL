```python
import pandas as pd
df = pd.read_csv("bing_news_shuffle.csv", encoding="UTF-8")
df
```

```python
import nltk
tokens = nltk.word_tokenize(df.iloc[0]['description'])
tokens
```

```python
tokens = [token.lower() for token in tokens if len(token) > 1]
```

## bigram과 trigram을 하는 이유는 이전 단어 A가 나왔을 때, 단어 B가 나올 수 있는 조건부 확률을 구하기 위해서이다
## 즉 단어 간의 의존관계를 확인하기 위함

```python
tokens_bigram = nltk.bigrams(tokens)
list(tokens_bigram)
```

```python
tokens_trigram = nltk.trigrams(tokens)
list(tokens_trigram)
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words("english")
stop_words
```

## 불용사 제거

```python
tokens_clean = [token for token in tokens if not token in stop_words]
tokens_clean
```

## 단어들에 대한 품사 정의

```python
tokens_tagged = nltk.pos_tag(tokens_clean)
tokens_tagged
```

```python
for word, pos in tokens_tagged:
    print(f"word : {word},   pos : {pos}")
```

```python
tokens_noun = [word for word, pos in tokens_tagged if pos == "NN"]
tokens_noun
```

## 단어들을 기본형으로 변환하기

```python
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
```

```python
word_list = ["am", "are", "is"]

for word in word_list:
    word_lem = lem.lemmatize(word, "v")
    print(f"word : {word}, lemmatize : {word_lem}")
```

```python
word_list = ["cats", "cat", "pupples", "dog", "dogs", "apples"]
for word in word_list:
    print(f"word : {word}, lemmatize : {lem.lemmatize(word, 'n')}")
```

```python
word_list = ["better", "good", "worst", "bad"]
for word in word_list:
    print(f"word : {word}, lemmatize : {lem.lemmatize(word, 'a')}")
```

## 워드클라우드로 출력

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
wc = WordCloud(background_color="white", max_font_size=100, max_words=50, stopwords=STOPWORDS)
```

```python
text_data = df['description'].str.cat()
text_data
```

```python
wordcloud = wc.generate(text_data)
wordcloud
```

```python
plt.figure(figsize = (20, 20))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
```
